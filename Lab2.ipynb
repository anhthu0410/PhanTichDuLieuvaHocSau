```python
import pandas as pd
import numpy as np
```

```python
#Đọc dữ liệu đầu vào
data = pd.read_csv("processed_dulieuxettuyendaihoc.csv", index_col= "STT")
```

```python
# Phần 1
#Câu 1: Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
data.sort_values(by="DH1")
```

```python
#Câu 2: Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
data.sort_values(by=['GT','DH2'])
```

```python
#Câu 3: Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT
def get_mean(scores):
    return sum(scores)/len(scores)

result = data['DH1'].sum()
print('\nThe result of applying the function on the dataframe:\n')
print(result)

def std(g):
    return np.std(g)

def q1(g):
    return np.percentile(g, 25)

def q2(g):
    return np.percentile(g, 50)

def q3(g):
    return np.percentile(g, 75)
```

```python
pd.pivot_table(data=data, index='KT',values='DH1', aggfunc=['sum','mean','median','min','max',std,q1,q2,q3]) 
```

```python
#Câu 4: Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,Q2 và Q3 của DH1 theo KT và KV
pd.pivot_table(data=data, index=['KT','KV'],values='DH1', aggfunc=['sum','mean','median','min','max',std,q1,q2,q3]) 
```

```python
#Câu 2: Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1):
def classify_T1(score):
    if score < 5:
        return 'kém'
    elif score < 7:
        return 'trung bình'
    elif score < 8:
        return 'khá'
    else:
        return 'giỏi'
```

```python
data['phanlopt1'] = data['T1'].apply(classify_T1)
```

```python
#Câu 2: Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
data['phanlopt1'] = np.random.choice(['Group1', 'Group2', 'Group3'], size=len(data))

data.boxplot(column='T1', by='phanlopt1', grid=False)
```
